```python
import pandas as pd

reviews = pd.read_csv("../wine-reviews/winemag-data-130k-v2.csv", index_col=0)

reviews.head()
```

# 1.

Quels sont les critiques de vin les plus fréquentes dans l'ensemble des données ? Créez une `Series` dont l'index est la catégorie `taster_twitter_handle` de l'ensemble de données, et dont les valeurs comptent le nombre de critiques que chaque personne a rédigées.

```python
# Deux solutions
reviews_written = reviews.groupby("taster_twitter_handle").taster_twitter_handle.count()
reviews_written = reviews["taster_twitter_handle"].value_counts()
reviews_written
```

# 2.

Quel est le meilleur vin que je puisse acheter pour une somme donnée ? Créez une `Series` dont l'indice est le prix du vin et dont la valeur est le nombre maximum de points qu'un vin coûtant le même prix a obtenu dans une critique. Triez les valeurs par prix, en ordre croissant (de sorte que `4.0` dollars soient en haut et `3300.0` dollars en bas).

```python
best_rating_per_price = reviews.groupby("price").points.agg([max])
best_rating_per_price
```

# 3.

Quels sont les prix minimum et maximum pour chaque cépage (`variety`) de vin ? Créez un `DataFrame` dont l'indice est la catégorie `variety` de l'ensemble de données et dont les valeurs sont les valeurs `min` et `max` de celui-ci.

```python
price_extremes = reviews.groupby(["variety"]).price.agg([min, max])
print(type(price_extremes))
price_extremes
```

# 4.

Créer une `Series` dont l'indice est celui des "reviewers" et dont la valeur est la note moyenne de l'examen donné par ce "reviewer".

Astuce : vous aurez besoin des colonnes `taster_name` et `points`.

```python
# Plusieurs solutions possibles
reviews["taster_name"].value_counts()
reviewer_mean_ratings = reviews.groupby(["taster_name"]).points.mean()
reviewer_mean_ratings = reviews.groupby(["taster_name"]).points.agg("mean") # passer le mean en string "mean", "count", "std"
reviewer_mean_ratings
# aller plus loin
reviewer_mean_ratings = reviews.groupby(["taster_name"]).points.agg([min, "mean", max]).round(2)
reviewer_mean_ratings
```

Y a-t-il des différences significatives dans les notes moyennes attribuées par les différents "reviewers" ? Exécutez le code ci-dessous qui utilise la méthode `describe()` afin de voir un résumé de la fourchette des valeurs.

```python
reviewer_mean_ratings.describe()
```

# 5.

Quelle est la combinaison de pays et de variétés la plus courante ? Créez une `Series` dont l'indice est un `MultiIndex` de paires `{country, variety}`. Par exemple, un pinot noir produit aux États-Unis doit être associé à {"US", "Pinot Noir"}. Triez les valeurs de la série par ordre décroissant en fonction du nombre de vins.

```python
country_variety_counts = reviews.groupby(["country", "variety"]).winery.agg([len])
#country_variety_counts = country_variety_counts.reset_index()
#reviews.columns
country_variety_counts.sort_values(by ='len', ascending = False)
```
